# Step 1. Import the necessary libraries

```python
import pandas as pd
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')
print(users.head())
```

# Step 4: Discover the mean age per occupation

```python
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Mean Age per Occupation:\n", mean_age_per_occupation)
```

 # Step 5: Discover the Male ratio per occupation and sort it

```python
def male_ratio(group):
    return (group['gender'] == 'M').sum() / group['gender'].count()

male_ratio_per_occupation = users.groupby('occupation').apply(male_ratio).sort_values(ascending=False)
print("\nMale Ratio per Occupation (sorted):\n", male_ratio_per_occupation)
```

# Step 6: Calculate the minimum and maximum ages per occupation

```python
age_range_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nAge Range per Occupation:\n", age_range_per_occupation)
```

# Step 7: Calculate the mean age per combination of occupation and sex

```python
mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean Age per Occupation and Gender:\n", mean_age_per_combination)
```

## Step 8: Calculate the percentage of women and men per occupation

```python
def gender_percentage(group):
    total_count = group['gender'].count()
    female_count = (group['gender'] == 'F').sum()
    male_count = total_count - female_count
    return (female_count / total_count) * 100, (male_count / total_count) * 100

gender_percentage_per_occupation = users.groupby('occupation').apply(gender_percentage)
print("\nPercentage of Women and Men per Occupation:\n", gender_percentage_per_occupation)
```

```python
import pandas as pd

# Import the dataset
file_path = "C:\\Users\\satya\\Downloads\\wind.txt"
data = pd.read_csv(file_path, delim_whitespace=True)

# Combine year, month, and day columns into a proper datetime index
data['date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg(' '.join, axis=1))
data.set_index('date', inplace=True)

# Fix the year to be in the correct range
def fix_year(year):
    if year >= 90:
        return year + 1900
    else:
        return year + 2000
data['Yr'] = data['Yr'].apply(fix_year)

#Set the correct dates as the index
data.index = pd.to_datetime(data.index)
data.drop(['Yr', 'Mo', 'Dy'], axis=1, inplace=True)
data = data.astype(float)  # Convert the data to float

#Compute missing values per location
missing_values_per_location = data.isnull().sum()

#Compute total non-missing values
total_non_missing_values = data.count().sum()

#Calculate the mean windspeed over all locations and times
mean_windspeed = data.mean().mean()

#Create loc_stats DataFrame
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

# Create day_stats DataFrame
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})

# average windspeed in January for each location
january_avg_per_location = data[data.index.month == 1].mean()

#

# results
print("Missing values per location:")
print(missing_values_per_location)
print("\nTotal non-missing values:", total_non_missing_values)
print("\nMean windspeed over all locations and times:", mean_windspeed)
print("\nLocation statistics:")
print(loc_stats)
print("\nDay statistics:")
print(day_stats)
print("\nAverage windspeed in January for each location:")
print(january_avg_per_location)
```

```python
import pandas as pd

#Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

#'chipo' is assigned to the variable 'chipo'

#  first 10 entries
print("Step 4: First 10 entries in the dataset:\n", chipo.head(10))

# Number of observations in the dataset
num_observations = chipo.shape[0]
print("\nStep 5: Number of observations in the dataset:", num_observations)

# Number of columns in the dataset
num_columns = chipo.shape[1]
print("\nStep 6: Number of columns in the dataset:", num_columns)

# Print the name of all columns
print("\nStep 7: Name of all columns:\n", chipo.columns)

#  How is the dataset indexed?
print("\nStep 8: Dataset index:\n", chipo.index)

#  Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("\nStep 9: Most-ordered item:", most_ordered_item)

#  Number of items ordered for the most-ordered item
most_ordered_item_count = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("\nStep 10: Number of items ordered for the most-ordered item:", most_ordered_item_count)

#  Most ordered item in the choice_description column
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()
print("\nStep 11: Most ordered item in the choice_description column:", most_ordered_choice)

# Total number of items ordered
total_items_ordered = chipo['quantity'].sum()
print("\nStep 12: Total number of items ordered:", total_items_ordered)

# Convert item price to a float
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)

# Revenue for the period in the dataset
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("\nStep 14: Revenue for the period in the dataset:", revenue)

# Number of orders made in the period
num_orders = chipo['order_id'].nunique()
print("\nStep 15: Number of orders made in the period:", num_orders)

# Average revenue amount per order
avg_revenue_per_order = revenue / num_orders
print("\nStep 16: Average revenue amount per order:", avg_revenue_per_order)

# Number of different items sold
num_different_items = chipo['item_name'].nunique()
print("\nStep 17: Number of different items sold:", num_different_items)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import the data from the file
path = r"C:\Users\satya\Downloads\us-marriages-divorces-1867-2014.csv"
data = pd.read_csv(path)

# Filter data for the years 1900, 1950, and 2000
years_of_interest = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years_of_interest)]

# Set up the figure and axes
plt.figure(figsize=(10, 6))

# Vertical bar chart for marriages per capita
plt.bar(filtered_data['Year'], filtered_data['Marriages_per_1000'], label='Marriages per capita', color='skyblue', width=0.4)

# Vertical bar chart for divorces per capita (shifted by 0.4 for better visibility)
plt.bar(filtered_data['Year'] + 0.4, filtered_data['Divorces_per_1000'], label='Divorces per capita', color='salmon', width=0.4)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')

# Add legend
plt.legend()

# Adjust x-axis ticks and labels
plt.xticks(filtered_data['Year'] + 0.2, years_of_interest)

# Improve aesthetics
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


path = r"C:\Users\satya\Downloads\roman-emperor-reigns.csv"
data = pd.read_csv(path)

# Count the number of assassinations
assassinations_count = data['Cause_of_Death'].str.contains('assassinated', case=False).sum()

# Count the number of non-assassinations
non_assassinations_count = len(data) - assassinations_count

# Labels for the pie chart
labels = ['Assassinated', 'Not Assassinated']

# Data for the pie chart
sizes = [assassinations_count, non_assassinations_count]

# Plotting
plt.figure(figsize=(8, 8))

# pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

plt.axis('equal')

# Add title
plt.title('Fraction of Roman Emperors Assassinated')

# Show the pie chart
plt.show()
```
